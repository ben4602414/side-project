```python
#建立EXCEL
# excel存放的資料夾
file_dir="/Users/ben/Documents/華梵課程"
# 若資料夾不存在，建立指定資料夾
file_out=file_dir+"/20220426_農產品_最高價格.xlsx"
```

```python
#最高價格
high_df = pd.DataFrame()
df = pd.read_excel("20220426_農作物.xlsx")
# iterate 年份
for year in range(2009,2022):
    for agriculture in np.unique(df[["作物"]].values): # iterate 農作物
        #排除價格為0的資料 並 排除價格為空白的資料
        if df[(df.作物 == agriculture) & (df.年 == year) & (df.價格 != 0)].nlargest(1,"價格")[["價格"]].empty != True:
            Largest_price = df[(df.作物 == agriculture) & (df.年 == year) & (df.價格 != 0)].nlargest(1,"價格")[["價格"]]
            Largest_price = Largest_price.values[0][0]
            
            data = df[(df.作物 == agriculture) & (df.年 == year) & (df.價格 == Largest_price)][["作物","月份","價格"]]
            #df append df
            high_df = high_df.append(data)

    
    print("已完成%d年的資料" % (year))
    print("-------------------------------")
#改Title
high_df = high_df.rename(columns={"作物": "最高_作物", "月份": "最高_月份", "價格": "最高_價格"})
# 如果檔案不存在，建立檔案
if  os.path.exists(file_out) == False:
    high_df.to_excel(file_out,index=False)
print("匯出完成")
print("-------------------------------")
print(high_df)
```

```python
#平均價格
avg_price = {
    "平均_作物":[],
    "平均_年份":[],
    "平均_價格":[],
}
df = pd.read_excel("20220426_農作物.xlsx")
# iterate 年份
for year in range(2009,2022):
    for agriculture in np.unique(df[["作物"]].values): # iterate 農作物
        #排除價格為0的資料 並 排除價格為空白的資料
        if df[(df.作物 == agriculture) & (df.年 == year) & (df.價格 != 0)][["價格"]].empty != True:
            Average_price = df[(df.作物 == agriculture) & (df.年 == year) & (df.價格 != 0)][["價格"]].mean()
            Average_price = Average_price.values[0]
            
            avg_price["平均_作物"].append(agriculture)
            avg_price["平均_年份"].append(year)
            avg_price["平均_價格"].append("%.3f" % (Average_price))
    
    print("已完成%d年的資料" % (year))
    print("-------------------------------")
    
# 如果檔案不存在，建立檔案
if  os.path.exists(file_out) == False:
    df = pd.DataFrame(avg_price) 
    df.to_excel(file_out,index=False)
print("匯出完成")
print("-------------------------------")
print(pd.DataFrame(avg_price))

```
